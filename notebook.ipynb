```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Import dataset
energy_data = pd.read_csv("/workspaces/Unveiling-Energy-Trends-in-Renewable-Energy/Training_set_augmented.csv")
display(energy_data.head())
display(energy_data.columns)
display(energy_data.shape)
```

## Which regions are investing the most efficiently in renewables?

```python
# Compute Production Efficiency (kWh/USD) = Production (GWh) / Investments (USD)
energy_data['Production Efficiency (kWh/USD)'] = (
    energy_data['Production (GWh)'] / energy_data['Investments (USD)']
    ) * 1e6
```

### How does production efficiency change over time for each country?

```python
# Pivot table: rows = countries, columns = years
heatmap_data = energy_data.pivot_table(values='Production Efficiency (kWh/USD)',
                              index='Country',
                              columns='Year',
                              aggfunc='mean'
                        )

# heatmap_data.describe()

# Plot heatmap
ax = sns.heatmap(
    heatmap_data,
    cmap='YlOrBr',
    annot=False,
    cbar_kws={'label': 'Production Efficiency (kWh/USD)'}
);
# Set title
plt.title('Average Renewable Energy Production Efficiency by Country and Year', fontsize=16, pad=20)

# Move the colorbar just below the title
colorbar = ax.collections[0].colorbar
colorbar.ax.tick_params(labelsize=10)
colorbar.set_label('Production Efficiency (kWh/USD)', fontsize=12)

# Adjust layout so title and colorbar have space
plt.tight_layout(rect=[0, 0, 1, 0.95])  # top space reserved for title

plt.show()
```
